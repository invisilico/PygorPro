```python
# code from plot series generator

import os 
import numpy as np

parent = "Desktop/VMHvl Shell Inputs/#2_Dec23/"

day = "231222M#1048/"

date = "Dec2322"

cell = "D"

dirname = "C:/users/njana/"+parent+day+date+cell+"/"+date+cell+"_Waves"
print(dirname)

wave_list = os.listdir(dirname) # takes entire filename ex: Dec2322D_1_10_1_1.ibw

#finds series and sweep numbers in list

series = []
sweeps = []
for ele in wave_list:
    rev = ele[::-1]
    loc_1 = rev.find("_")+1
    loc_2 = rev.find("_", rev.find("_") + 1)+1
    series += [rev[loc_2:loc_2+2][::-1].replace("_","")]
    sweeps += [rev[loc_1:loc_1+2][::-1].replace("_","")]
    
series = np.array(series[::2]) # takes every second series number because every

# finds unique series numbers, and how many there are

ser_nums, sweep_count = np.unique(series, return_counts=True)

# counter only to find how many and how many missed

dp_counter = 0
sp_counter = 0
train_counter = 0
printed_counter = 0

#prints plotting functions for each sweep thing

print("Plotting commands for cell "+cell+"\n")

for i in range(len(sweep_count)):
    ser_num = ser_nums[i]
    if sweep_count[i] == 10:
        print("PlotSeries("+ser_num+",1,10,2,0,1,\"TwoPulse_"+ser_num+"\")")
        print("Legend/K/N=text0")
        dp_counter += 1
        printed_counter += 1
    if sweep_count[i] == 3:
        print("PlotSeries("+ser_num+",1,3,2,0,1,\"TestPulse_"+ser_num+"\")")
        print("Legend/K/N=text0")
        sp_counter += 1
        printed_counter += 1
    if sweep_count[i] == 1:
        print("PlotSeries("+ser_num+",1,1,2,0,1,\"Train20P_"+ser_num+"\")")
        print("Legend/K/N=text0")
        train_counter += 1
        printed_counter += 1
        
#print the counter to see if any were missed
        
print("\nTotal number of Seriess: "+str(len(ser_nums)))
print(  "Total number of 2 Pulse: "+str(dp_counter))
print(  "Total number of 1 Pulse: "+str(sp_counter))
print(  "Total number of Train20: "+str(train_counter))
print("\nTotal Missed serieses  : "+str(len(ser_nums)-printed_counter))
print("\n"+str(ser_nums))
```
