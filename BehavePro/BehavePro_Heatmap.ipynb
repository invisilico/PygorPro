Generates Heatmaps and Trajectory Plots for sessions

```python
#import everything and set dpi here

import os
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100 #set dpi for viewing figs within the notebook
import scipy.stats as stats
from tqdm import tqdm
```

## Infotable

```python
directory = r"C:/Users/njana/Desktop/Behavior/Esr1ChronosStim/Day 4/" # include / at end

infofile = directory+"deets.csv"
infotable = pd.read_csv(infofile)

infotable
```

### Plot HEATMAPS

```python
''' Plot Heatmaps''' 

#ilst = [0]
#for i in ilst:
for i in tqdm(range(infotable.shape[0])):
    #get the data
    BlindID = infotable["BlindID"][i]
    simbarawdata = pd.read_csv(directory+infotable.name[i]+".csv", header = 0)

    edge = 10
    cmap = 'inferno'
    vmax = None

    plt.subplots(2,1)
    plt.subplot(2,1,1)
    sns.set_style("white")
    plt.hist2d(simbarawdata['Center_1_x'],simbarawdata['Center_1_y'], range = [[0,640],[0,400]], bins = [int(640/edge),int(400/edge)], cmap = cmap, vmax = vmax)
    plt.gca().set_aspect(1)
    plt.xlim(0,640)
    plt.gca().set_xticks([])
    plt.gca().set_yticks([])
    plt.ylim(0,400)
    sns.despine(bottom = True, left = True)
    plt.colorbar(fraction = 0.02)
    plt.title("Heatmap of Resident centroid", loc = 'left')

    plt.subplot(2,1,2)
    sns.set_style("white")
    plt.hist2d(simbarawdata['Center_2_x'],simbarawdata['Center_2_y'], range = [[0,640],[0,400]], bins = [int(640/edge),int(400/edge)], cmap = cmap, vmax = vmax)
    plt.gca().set_aspect(1)
    plt.xlim(0,640)
    plt.gca().set_xticks([])
    plt.gca().set_yticks([])
    plt.ylim(0,400)
    sns.despine(bottom = True, left = True)
    plt.colorbar(location = 'right',fraction = 0.02)
    plt.title("Heatmap of Intruder centroid", loc = 'left')
    plt.gcf().suptitle("Resident "+BlindID+": Day "+day+" Trial "+trial, ha = 'left', x = 0.125, y = 0.925, fontsize = 14, fontweight = "bold")
    plt.gcf().set_size_inches(6,8)
    plt.savefig(directory+BlindID+"_D"+day+"Tr"+trial+"_Heatmap.svg", dpi = 600)
    plt.savefig(directory+BlindID+"_D"+day+"Tr"+trial+"_Heatmap.png", dpi = 600)
    plt.clf()    

    sns.set_style("white")
    plt.plot(simbarawdata['Center_1_x'],simbarawdata['Center_1_y'], linewidth = 0.5, color = 'darkblue')
    plt.plot(simbarawdata['Center_2_x'],simbarawdata['Center_2_y'], linewidth = 0.1, color = 'darkred')
    plt.gca().set_aspect(1)
    plt.xlim(0,640)
    plt.gca().set_xticks([])
    plt.gca().set_yticks([])
    plt.ylim(0,400)
    sns.despine(bottom = True, left = True)
    plt.title("Trajectory of Resident and Intruder Centroids")
    plt.gcf().suptitle("Resident "+BlindID+": Day "+day+" Trial "+trial, ha = 'center', x = 0.5, y = 0.85, fontsize = 14, fontweight = "bold")
    plt.gcf().set_size_inches(10,8)
    plt.savefig(directory+BlindID+"_D"+day+"Tr"+trial+"_Trajectory.svg", dpi = 600)
    plt.savefig(directory+BlindID+"_D"+day+"Tr"+trial+"_Trajectory.png", dpi = 600)
    plt.clf()
```
